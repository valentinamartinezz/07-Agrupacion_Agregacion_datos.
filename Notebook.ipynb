```python
import pandas as pd
```

```python
phone = pd.read_csv('Data/smartphones.csv')
```

### 1.Agrupaci√≥n por Marca y Promedio de Precios Utiliza groupby para agrupar los smartphones por marca y calcula el precio promedio de cada una.

```python
marcas_precio = phone.groupby('Brand')['Final Price'].sum()
```

```python
marcas_precio
```

### 2.Conteo de Modelos por Marca Emplea groupby para contar la cantidad de modelos diferentes que tiene cada marca.

```python
modelos_por_marca=phone.groupby('Brand')['Model'].nunique()
```

```python
modelos_por_marca
```

### 3.RAM Promedio por Marca Usa groupby para determinar la cantidad promedio de RAM para cada marca de smartphone.

```python
ram_por_marca=phone.groupby('Brand')['RAM'].mean()
```

```python
ram_por_marca
```

### 4.Tabla Pivot de Precios M√°ximos por Color Crea una tabla pivot que muestre el precio m√°ximo de los smartphones para cada color.


```python
phone.pivot_table(values='Final Price', index='Color', aggfunc='max')
```

### 5.Distribuci√≥n de Almacenamiento por Marca Con groupby, calcula la distribuci√≥n de la capacidad de almacenamiento para cada marca.


```python
almacenimiento_por_marca=phone.groupby('Brand')['Storage'].value_counts()
```

```python
almacenimiento_por_marca
```

### 6.Precio Promedio por Combinaci√≥n de RAM y Almacenamiento Genera una tabla pivot que muestre el precio promedio para cada combinaci√≥n √∫nica de RAM y almacenamiento.

```python
phone.pivot_table(values='Final Price', index=['RAM','Storage'], aggfunc='mean')
```

### 7.Cantidad de Smartphones Libres por Marca Utiliza groupby para contar cu√°ntos smartphones de cada marca est√°n libres de contratos con compa√±√≠as celulares.

```python
smart_por_marca=phone.groupby('Brand')['Smartphone'].count()
```

```python
smart_por_marca
```

### 8.Precio M√≠nimo, M√°ximo y Promedio por Marca Crea una tabla pivot que resuma el precio m√≠nimo, m√°ximo y promedio para cada marca de smartphone.

```python
phone.pivot_table(values='Final Price', index='Brand', aggfunc=['min','max','mean'])
```

### 9.Smartphones por Rango de Precios Emplea groupby para agrupar los smartphones en rangos de precios y cuenta cu√°ntos hay en cada rango.

```python
phone ['Rang Price'] = pd.cut(phone['Final Price'],bins= [0,300,600,900,1200,1500,1800,2100,2400])
```

```python
phone.groupby('Rang Price').size()
```

### 10.Comparaci√≥n de Precios entre Smartphones Libres y con Contrato Genera una tabla pivot que compare el precio promedio de los smartphones libres (Free = Yes) versus los que est√°n atados a contratos (Free = No).

```python
phone.pivot_table(values='Final Price', index='Free', aggfunc='mean')
```
